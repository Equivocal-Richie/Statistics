```python
#Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# DataFrame
data = {
    'CustomerID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Age': [28, 35, 42, 29, 55, 30, 45, 22],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Purchase_Amount': [150.50, 200.75, 50.00, 320.10, 88.99, 120.00, 250.00, 75.50],
    'Product_Category': ['Electronics', 'Clothing', 'Electronics', 'Home Goods', 'Clothing', 'Electronics', 'Home Goods', 'Clothing'],
    'Satisfaction_Rating': ['Good', 'Excellent', 'Poor', 'Excellent', 'Good', 'Poor', 'Good', 'Excellent']
}
df = pd.DataFrame(data)
```

```python
# Let's view the first 5 rows
print(df.head())
```

```python
print("\n--- Visualizing Categorical Data ---\n")

# Bar Chart for 'Gender'
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gender', palette='viridis', hue='Gender', legend=False) # countplot is great for categorical counts
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Bar Chart for 'Product_Category'
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Product_Category', palette='mako', hue='Product_Category', legend=False)
plt.title('Distribution of Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()

# Pie Chart for 'Gender' (only for small number of categories)
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Gender')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
print("\n--- Visualizing Numerical Data ---\n")

# Histogram for 'Age'
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', kde=True, bins=5, color='skyblue') # kde=True adds the density curve
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Histogram for 'Purchase_Amount'
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Purchase_Amount', kde=True, bins=7, color='lightcoral')
plt.title('Distribution of Purchase Amount')
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')
plt.show()


# Box Plot for 'Age'
plt.figure(figsize=(6, 5))
sns.boxplot(data=df, y='Age', color='lightgreen') # For a single variable, use y-axis
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()

# Box Plot for 'Purchase_Amount'
plt.figure(figsize=(6, 5))
sns.boxplot(data=df, y='Purchase_Amount', color='gold')
plt.title('Box Plot of Purchase Amount')
plt.ylabel('Purchase Amount')
plt.show()

# Box Plot for 'Purchase_Amount' by 'Product_Category' (comparison across categories)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Product_Category', y='Purchase_Amount', palette='coolwarm', hue='Product_Category', legend=False)
plt.title('Purchase Amount Distribution by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Purchase Amount')
plt.show()


# Scatter Plot: Age vs Purchase Amount
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Purchase_Amount', hue='Gender', size='Purchase_Amount', sizes=(20, 400), alpha=0.7)
plt.title('Age vs. Purchase Amount')
plt.xlabel('Age')
plt.ylabel('Purchase Amount')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
