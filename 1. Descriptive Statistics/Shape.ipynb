```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Add the dataFrame
data = {
    'CustomerID': [101, 102, 103, 104, 105],
    'Age': [28, 35, 42, 29, 55],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Purchase_Amount': [150.50, 200.75, 50.00, 320.10, 88.99],
    'Product_Category': ['Electronics', 'Clothing', 'Electronics', 'Home Goods', 'Clothing'],
    'Satisfaction_Rating': ['Good', 'Excellent', 'Poor', 'Excellent', 'Good']
}
df = pd.DataFrame(data)

# Let's view the first 5 rows
print("\n First 5 rows of the DataFrame: \n")
print(df.head())
```

# **1. Skewness and Kurtois**

```python
print("\n--- Measures of Shape (Skewness & Kurtosis) ---")

# For numerical columns
numerical_cols = ['Age', 'Purchase_Amount']

for col in numerical_cols:
    print(f"\n--- {col} ---")

    # Skewness
    data_skewness = df[col].skew()
    print(f"Skewness: {data_skewness:.2f}")

    # Kurtosis (Fisher's Excess Kurtosis)
    data_kurtosis = df[col].kurtosis()
    print(f"Kurtosis (Fisher): {data_kurtosis:.2f}")
```

## **Interpreting the output**

### **1. Skewness**

Both `Age` and `Purchase Amount` columns are positively skewed (asymmetrical) - right skewed.


### **2. Kurtois**

Both `Age` and `Purchase Amount` columns are leptokurtic (positive values) meaning the distribution has sharper peaks with thicker tails.


## **Effects of outliers**

```python
# Let's add the outlier again to see the effect
df_with_outlier = df.copy()
df_with_outlier.loc[5] = {'CustomerID': 106, 'Age': 200, 'Gender': 'Female', 'Purchase_Amount': 10000, 'Product_Category': 'Electronics', 'Satisfaction_Rating': 'Excellent'}

print("\n--- Measures of Shape with Outlier ---")
for col in numerical_cols:
     print(f"\n--- {col} (with outlier) ---")
     print(f"Skewness: {df_with_outlier[col].skew():.2f}") # Skewness is sensitive to outliers
     print(f"Kurtosis (Fisher): {df_with_outlier[col].kurtosis():.2f}") # Kurtosis is also sensitive to outliers
```

# **2. Visualizing Shape**

```python
# --- Visualizing Shape ---
# A great way to see the shape is with a histogram or density plot

print("\n--- Visualizing Data Shape (Histograms & Density Plots) ---")

# Create histograms and density plots for the original data
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True) # kde=True adds the density curve
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency / Density')
    plt.show()
```

## **Interpreting the output**

Both distributions are right skewed; typically the mean is greater than the median.

```python
# Create histograms and density plots for the data with outliers
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df_with_outlier[col], kde=True)
    plt.title(f'Distribution of {col} (with Outlier)')
    plt.xlabel(col)
    plt.ylabel('Frequency / Density')
    plt.show()
```
